```python
# PIL package

import matplotlib.pyplot as plt
import os
%matplotlib inline

from PIL import Image
pil_iimage = Image.open('image/dog.jpg')
print('image type : ', type(pil_iimage))

plt.figure(figsize = (10, 10))
plt.imshow(pil_iimage)
```

```python
# skimage
# skimage imread() image numpy  

import matplotlib.pyplot as plt
from skimage import io

sk_image = io.imread('image/dog.jpg')
print('skimage type : ', type(sk_image), 'skimage shape : ', sk_image.shape)

plt.figure(figsize = (10, 10))
plt.imshow(sk_image)
```

```python
# opencv
# opencv imread()   RGB image BGR  numpy  
# imwrite()  BGR  image      RGB  

import cv2
import matplotlib.pyplot as plt

cv2_image = cv2.imread('image/dog.jpg')
cv2.imwrite('image/dog_1.jpg', cv2_image)
print('image type :', type(cv2_image), 'image shapge : ', cv2_image.shape)

plt.figure(figsize = (10, 10))
img = plt.imread('image/dog_1.jpg')
plt.imshow(img)
```

```python
# BGR imaage

import cv2
import matplotlib.pyplot as plt

bgr_image = cv2.imread('image/dog.jpg')

plt.figure(figsize = (10, 10))
plt.imshow(bgr_image)
plt.show()
```
