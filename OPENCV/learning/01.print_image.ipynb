##### RGB & BGR

opencv에서 cv2.imread를 통해 이미지를 읽을 때 BGR로 저장한다. 하지만, matplotlib는 RGB로 저장하므로, cv2.COLOR_BGR2RGB으로 변환해야한다.

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('image/img.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # RGB -> BGR

fig = plt.figure(figsize = (10, 10))
rows = 2
cols = 1

ax1 = fig.add_subplot(rows, cols, 1)
ax1.set_title('BGR image')
plt.imshow(img)

ax2 = fig.add_subplot(rows, cols, 2)
ax2.set_title('RGB image')
plt.imshow(img_rgb)

plt.show()
```

##### 읽기 옵션 (1)

> cv2.IMREAD_COLOR : 컬러 이미지, 투명 영역은 무시(default)

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('image/img.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_color = cv2.imread('image/img.jpg', cv2.IMREAD_COLOR) # cv2.IMREAD_COLOR : 컬러 이미지, 투명 영역은 무시(기본값)
img_color_rgb = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)


fig = plt.figure(figsize = (20, 20))
rows = 1
cols = 2

ax1 = fig.add_subplot(rows, cols, 1)
ax1.set_title('original image')
plt.imshow(img_rgb)

ax2 = fig.add_subplot(rows, cols, 2)
ax2.set_title('color image')
plt.imshow(img_color_rgb)

plt.show()
```

##### 읽기 옵션 (2)

> cv2.IMREAD_GRAYSCALE : 흑백이미지

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('image/img.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_gray = cv2.imread('image/img.jpg', cv2.IMREAD_GRAYSCALE) # cv2.IMREAD_GRAYSCALE : 흑백 이미지
img_gray_rgb = cv2.cvtColor(img_gray, cv2.COLOR_BGR2RGB)


fig = plt.figure(figsize = (20, 20))
rows = 1
cols = 2

ax1 = fig.add_subplot(rows, cols, 1)
ax1.set_title('original image')
plt.imshow(img_rgb)

ax2 = fig.add_subplot(rows, cols, 2)
ax2.set_title('gray image')
plt.imshow(img_gray_rgb)

plt.show()
```

##### 읽기 옵션 (3)

> cv2.IMREAD_UNCHANGED : 투명 영역까지 포함

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('image/img.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_unchanged = cv2.imread('image/img.jpg', cv2.IMREAD_UNCHANGED) # cv2.IMREAD_UNCHANGED : 투명 영역까지 포함
img_unchanged_rgb = cv2.cvtColor(img_unchanged, cv2.COLOR_BGR2RGB)


fig = plt.figure(figsize = (20, 20))
rows = 1
cols = 2

ax1 = fig.add_subplot(rows, cols, 1)
ax1.set_title('original image')
plt.imshow(img_rgb)

ax2 = fig.add_subplot(rows, cols, 2)
ax2.set_title('unchanged image')
plt.imshow(img_unchanged_rgb)

plt.show()
```
