##### 이미지

고정 크기로 설정

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('image/img.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
dst = cv2.resize(img_rgb, (400, 500)) # width, height 고정크기

fig = plt.figure(figsize = (10, 10))
rows = 1
cols = 2

ax1 = fig.add_subplot(rows, cols, 1)
ax1.set_title('original image')
plt.imshow(img_rgb)

ax1 = fig.add_subplot(rows, cols, 2)
ax1.set_title('risezed image')
plt.imshow(dst)

plt.show()
```

비율로 설정

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('image/img.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
dst = cv2.resize(img_rgb, None, fx=0.5, fy=0.5) # x, y 비율 정의 (0.5 배로 축소)

fig = plt.figure(figsize = (20, 20))
rows = 1
cols = 2

ax1 = fig.add_subplot(rows, cols, 1)
ax1.set_title('original image')
plt.imshow(img_rgb)

ax1 = fig.add_subplot(rows, cols, 2)
ax1.set_title('resized image')
plt.imshow(dst)

plt.show()
```

##### 보간법

> cv2.INTER_AREA : 크기 줄일 때 사용

> cv2.INTER_CUBIC : 크기 늘릴 때 사용 (속도 느림, 퀄리티 좋음)

> cv2.INTER_LINEAR : 크기 늘릴 때 사용 (기본값)

보간법 적용하여 축소

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('image/img.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
dst = cv2.resize(img_rgb, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_AREA) # x, y 비율 정의 (0.5 배로 축소)

fig = plt.figure(figsize = (20, 20))
rows = 1
cols = 2

ax1 = fig.add_subplot(rows, cols, 1)
ax1.set_title('original image')
plt.imshow(img_rgb)

ax1 = fig.add_subplot(rows, cols, 2)
ax1.set_title('resized image')
plt.imshow(dst)

plt.show()
```

보간법 적용하여 확대

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('image/img.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
dst = cv2.resize(img_rgb, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_CUBIC) # x, y 비율 정의 (1.5 배로 축소)

fig = plt.figure(figsize = (20, 20))
rows = 1
cols = 2

ax1 = fig.add_subplot(rows, cols, 1)
ax1.set_title('original image')
plt.imshow(img_rgb)

ax1 = fig.add_subplot(rows, cols, 2)
ax1.set_title('resized image')
plt.imshow(dst)

plt.show()
```

##### 동영상

고정 크기로 설정

```python
import cv2

cap = cv2.VideoCapture('video/video.mp4')

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame_resized = cv2.resize(frame, (400, 500))

    cv2.imshow('video', frame_resized)
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
