##### Save image

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('image/img.jpg', cv2.IMREAD_GRAYSCALE) #   
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure()
plt.imshow(img_rgb)
plt.show()

result = cv2.imwrite('image/img_save.jpg', img)
print(result)
```

##### Save video

```python
import cv2
cap = cv2.VideoCapture('video/video.mp4')

#   ()
fourcc = cv2.VideoWriter_fourcc(*'DIVX')

#  , FPS 
width = round(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) # width, height  ,       round 
height = round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS) #   2  cap.get(cv2.CAP_PROP_FPS) * 2

# VideoWriter  
# cv2.VideoWriter( , , FPS, (width, height))
out = cv2.VideoWriter('video/output.avi', fourcc, fps, (width, height))


while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    out.write(frame) #    (x)

    cv2.imshow('video', frame)
    if cv2.waitKey(1) == ord('q'):
        break

out.release() #  
cap.release()
cv2.destroyAllWindows()
```
