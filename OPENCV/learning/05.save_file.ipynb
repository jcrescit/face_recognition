##### Save image

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('image/img.jpg', cv2.IMREAD_GRAYSCALE) # 흑백으로 이미지 불러오기
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure()
plt.imshow(img_rgb)
plt.show()

result = cv2.imwrite('image/img_save.jpg', img)
print(result)
```

##### Save video

```python
import cv2
cap = cv2.VideoCapture('video/video.mp4')

# 코텍 정의 (형태)
fourcc = cv2.VideoWriter_fourcc(*'DIVX')

# 프레임의 크기, FPS 정의
width = round(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) # width, height는 정수값을 받아야하는데, 실수값을 받으면 에러가 발생할 수 있으므로 round로 처리
height = round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS) # 영상 속도를 2배로 변경하려면 cap.get(cv2.CAP_PROP_FPS) * 2

# VideoWriter 객체 제작
# cv2.VideoWriter(저장 파일명, 코덱, FPS, 크기(width, height))
out = cv2.VideoWriter('video/output.avi', fourcc, fps, (width, height))


while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    out.write(frame) # 영상 데이터만 저장 (소리x)

    cv2.imshow('video', frame)
    if cv2.waitKey(1) == ord('q'):
        break

out.release() # 자원 해제
cap.release()
cv2.destroyAllWindows()
```
