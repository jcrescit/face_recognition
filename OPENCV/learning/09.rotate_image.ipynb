#### 09. Rotate image

시계 방향 90도 회전

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('image/img.jpg')

rotate_90 = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE) # 시계 방향 90도 회전

fig = plt.figure(figsize = (10, 10))
rows = 1
cols = 2

ax1 = fig.add_subplot(rows, cols, 1)
ax1.set_title('original image')
plt.imshow(img)

ax2 = fig.add_subplot(rows, cols, 2)
ax2.set_title('rotate image')
plt.imshow(rotate_90)

plt.show()
```

시계 반대 방향 90도 회전 (시계 방향 270도 회전)

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('image/img.jpg')
rotate_270 = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE) # 시계 방향 270도 회전

fig = plt.figure(figsize = (10, 10))
rows = 1
cols = 2

ax1 = fig.add_subplot(rows, cols, 1)
ax1.set_title('original image')
plt.imshow(img)

ax2 = fig.add_subplot(rows, cols, 2)
ax2.set_title('rotate image')
plt.imshow(rotate_270)

plt.show()
```
