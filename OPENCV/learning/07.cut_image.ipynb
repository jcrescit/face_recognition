#### 07. Cut image

영역을 잘라서 새로운 윈도우 (창)에 표시

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('image/img.jpg')
# img.shape # (390, 640, 3)

crop = img[100:200, 200:400] # 세로 기준 100:200, 가로 기준 200:400 까지 자름

fig = plt.figure(figsize = (15, 15))
rows = 1
cols = 2

ax1 = fig.add_subplot(rows, cols, 1)
ax1.set_title('original image')
plt.imshow(img)

ax1 = fig.add_subplot(rows, cols, 2)
ax1.set_title('cropped image')
plt.imshow(crop)

plt.show()
```

영역을 잘라서 기존 윈도우에 표시

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('image/img.jpg')

crop = img[100:200, 200:400] # 세로 기준 100:200, 가로 기준 200:400 까지 자름
img[100 : 200, 400:600] = crop # 원본 이미지 100:200, 400:600 위치에 crop을 넣음으로써 하나의 이미지로 나타냄

plt.figure()
plt.imshow(img)
plt.show()
```
