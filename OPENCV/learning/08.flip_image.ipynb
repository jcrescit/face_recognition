#### 08.Flip image

##### 좌우 대칭

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('image/img.jpg')
flip_horizontal = cv2.flip(img, 1) # flip code > 0 : 좌우 대칭 Horizontal

fig = plt.figure(figsize = (10, 10))
rows = 1
cols = 2

ax1 = fig.add_subplot(rows, cols, 1)
ax1.set_title('original image')
plt.imshow(img)

ax2 = fig.add_subplot(rows, cols, 2)
ax2.set_title('flipped image')
plt.imshow(flip_horizontal)

plt.show()
```

##### 상하 대칭

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('image/img.jpg')
flip_vertical = cv2.flip(img, 0) # flip code = 0 : 상하 대칭 Vertical

fig = plt.figure(figsize = (20, 20)) # rows * cols matrix
rows = 1
cols = 3

ax1 = fig.add_subplot(rows, cols, 1)
ax1.set_title('original image')
plt.imshow(img)

ax2 = fig.add_subplot(rows, cols, 2)
ax2.set_title('flipped image')
plt.imshow(flip_vertical)

plt.show()
```

##### 상하좌우 대칭

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('image/img.jpg')
flip_both = cv2.flip(img, -1) # flip code < 0 : 상하좌우 대칭

fig = plt.figure(figsize = (20, 20)) # rows * cols matrix
rows = 1
cols = 3

ax1 = fig.add_subplot(rows, cols, 1)
ax1.set_title('original image')
plt.imshow(img)

ax2 = fig.add_subplot(rows, cols, 2)
ax2.set_title('flipped image')
plt.imshow(flip_both)

plt.show()
```
