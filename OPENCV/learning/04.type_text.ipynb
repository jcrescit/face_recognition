#### 4. Text

> cv2.putText(이미지, 텍스트 내용, 시작 위치, 폰트 종류, 크기, 색깔, 두께)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((300, 400, 3), dtype = np.uint8)

SCALE = 1 # 크기
COLOR = (255, 255, 255) # White
THICKNESS = 1 # 두께

cv2.putText(img, "Simplex", (20, 50), cv2.FONT_HERSHEY_SIMPLEX, SCALE, COLOR, THICKNESS) # 보통 크기의 산 셰리프(sans-serif) 글꼴
cv2.putText(img, "Plain", (20,100), cv2.FONT_HERSHEY_PLAIN, SCALE, COLOR, THICKNESS) # 작은 크기의 산 세리프 글꼴
cv2.putText(img, "Script Simplex", (20, 150), cv2.FONT_HERSHEY_SCRIPT_SIMPLEX, SCALE, COLOR, THICKNESS) # 필기체 스타일 글꼴
cv2.putText(img, "Triplex", (20, 200), cv2.FONT_HERSHEY_TRIPLEX, SCALE, COLOR, THICKNESS) # 보통 크기의 산 세리프 글꼴
cv2.putText(img, "Italic", (20, 250), cv2.FONT_HERSHEY_TRIPLEX | cv2.FONT_ITALIC, SCALE, COLOR, THICKNESS) # 기울임(이탤릭체)

plt.figure(figsize = (10, 10))
plt.imshow(img)
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# PIL (Python Image Library)
from PIL import ImageFont, ImageDraw, Image

def myPutText(src, text, pos, font_size, font_color):
    img_pil = Image.fromarray(src)
    draw = ImageDraw.Draw(img_pil)
    font = ImageFont.truetype('fonts/gulim.ttc', font_size)
    draw.text(pos, text, font = font, fill = font_color)
    return np.array(img_pil)


img = np.zeros((300, 470, 3), dtype = np.uint8)

FONT_SIZE = 30
COLOR = (255, 255, 255) # White

img = myPutText(img, "꿈을 없고요 그냥 놀고 싶습니다", (20, 50), FONT_SIZE, COLOR)

plt.figure(figsize = (10, 10))
plt.imshow(img)
plt.show()
```
