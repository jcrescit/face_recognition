#### 11. Image transformation (Blur)

average blurring

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('image/img.jpg')

#    
kernel_1 = np.ones((3, 3))/3**2 # 3x3   
kernel_2 = np.ones((5,5))/5**2 # 5x5   

#  
blured_1 = cv2.filter2D(img, -1, kernel_1)
blured_2 = cv2.filter2D(img, -1, kernel_2)

fig = plt.figure(figsize = (20, 20))
rows = 1
cols = 3

ax1 = fig.add_subplot(rows, cols, 1)
ax1.set_title('origin image')
plt.imshow(img)

ax2 = fig.add_subplot(rows, cols, 2)
ax2.set_title('3x3 average filter')
plt.imshow(blured_1)

ax3 = fig.add_subplot(rows, cols, 3)
ax3.set_title('5x5 average filter')
plt.imshow(blured_2)

plt.show()
```

blur function

```python
import cv2
import numpy as np

file_name = 'image/img.jpg'
img = cv2.imread(file_name)

# blur()    ---
blur1 = cv2.blur(img, (10,10))
# boxFilter()    ---
blur2 = cv2.boxFilter(img, -1, (10,10))

#  
merged = np.hstack( (img, blur1, blur2))
cv2.imshow('blur', merged)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('image/img.jpg')

# blur()  
blur1 = cv2.blur(img, (10,10))
# boxFilter()   
blur2 = cv2.boxFilter(img, -1, (10,10))

fig = plt.figure(figsize = (20, 20))
rows = 1
cols = 3

ax1 = fig.add_subplot(rows, cols, 1)
ax1.set_title('original image')
plt.imshow(img)

ax2 = fig.add_subplot(rows, cols, 2)
ax2.set_title('blur() function image')
plt.imshow(blur1)

ax3 = fig.add_subplot(rows, cols, 3)
ax3.set_title('boxFilter() function image')
plt.imshow(blur2)

plt.show()
```

Gaussian Blurring

> cv2.GaussianBlur(src, ksize, sigmaX, sigmaY, borderType)

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('image/img.jpg')

# (3, 3) (5, 5) (7, 7)
kernel_3 = cv2.GaussianBlur(img, (3, 3), 0)
kernel_5 = cv2.GaussianBlur(img, (5, 5), 0)
kernel_7 = cv2.GaussianBlur(img, (7, 7), 0)

fig = plt.figure(figsize = (10, 10))
rows = 2
cols = 2

ax1 = fig.add_subplot(rows, cols, 1)
ax1.set_title('original image')
plt.imshow(img)

ax2 = fig.add_subplot(rows, cols, 2)
ax2.set_title('blur image_1')
plt.imshow(kernel_3)

ax2 = fig.add_subplot(rows, cols, 3)
ax2.set_title('blur image_2')
plt.imshow(kernel_5)

ax2 = fig.add_subplot(rows, cols, 4)
ax2.set_title('blur image_3')
plt.imshow(kernel_7)

plt.show()
```

```python
import cv2
img = cv2.imread('image/img.jpg')

sigma_1 = cv2.GaussianBlur(img, (0, 0), 1) # sigmaX :   x  
sigma_2 = cv2.GaussianBlur(img, (0, 0), 2)
sigma_3 = cv2.GaussianBlur(img, (0, 0), 3)

fig = plt.figure(figsize = (10, 10))
rows = 2
cols = 2

ax1 = fig.add_subplot(rows, cols, 1)
ax1.set_title('original image')
plt.imshow(img)

ax2 = fig.add_subplot(rows, cols, 2)
ax2.set_title('blur image_1')
plt.imshow(sigma_1)

ax2 = fig.add_subplot(rows, cols, 3)
ax2.set_title('blur image_2')
plt.imshow(sigma_2)

ax2 = fig.add_subplot(rows, cols, 4)
ax2.set_title('blur image_3')
plt.imshow(sigma_3)

plt.show()
```
