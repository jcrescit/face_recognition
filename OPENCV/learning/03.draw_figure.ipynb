##### 직선

직선의 종류 (line type)

> cv2.line(위치, 시작점, 끝점, 색깔, 두께, 선종류)

```python
# BGR

import cv2
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((220, 220, 3), dtype = np.uint8)

COLOR = (255, 255, 0)
THICKNESS = 2

cv2.line(img, (50, 100), (200, 20), COLOR, THICKNESS, cv2.LINE_4) # cv2.LINE_4 : 상하좌우 4방향으로 연결된 선
cv2.line(img, (50, 150), (200, 70), COLOR, THICKNESS, cv2.LINE_8) # cv2.LINE_8 : 대각선을 포함한 8방향으로 연결된 선 (default)
cv2.line(img, (50, 200), (200, 120), COLOR, THICKNESS, cv2.LINE_AA) # cv2.LINE_AA : 부드러운 선

plt.figure(figsize = (10, 10))
plt.imshow(img)
plt.show()
```

##### 원

> cv2.circle(이미지, 위치, 중심, 반지름, 색깔, 두께, 선 종류)

##### 호, 타원

> cv2.ellipse(이미지, 위치, 중심, rotate angle, start angle, end angle, color, thickness)

```python
# BGR

import cv2
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((680, 1040, 3), dtype = np.uint8)

# 원
cv2.circle(img, (200, 100), 50, (255, 0, 0)) # 중심 (200, 100), 반지름 50, red, 두께 default 1
cv2.circle(img, (400, 100), 70, (255, 255, 0), 5, cv2.LINE_AA) # 중심 (400, 100), 반지름 70, yellow, 두께 5, 부드러운 곡선
cv2.circle(img, (600, 100), 80, (255, 255, 255), cv2.FILLED, cv2.LINE_AA) # 중심 (600, 100), 반지름 80, white, 내부채움, 부드러운 곡선

# 반원
cv2.ellipse(img, (200, 300), (60, 60), 0, 0, 360, (0, 255, 0), 5) # 중심 (200, 300), 반지름 60, green 원
cv2.ellipse(img, (400, 300), (60, 60), 0, 0, 180, (0, 255, 0), 5) # 중심 (400, 300), 반지름 60, , green 윗 반원
cv2.ellipse(img, (600, 300), (60, 60), 0, 180, 360, (0, 255, 0), 5) # 중심 (600, 300), 반지름 60, green 아래 반원

# 타원
cv2.ellipse(img, (200, 500), (40, 80), 0, 0, 360, (0, 0, 255), 5) # 중심 (200, 500), 반지름 (40, 80), blue 타원
cv2.ellipse(img, (400, 500), (100, 50), 0, 0, 360, (0, 0, 255), 5) # 중심 (400, 500), 반지름 (100, 50), blue 타원
cv2.ellipse(img, (600, 500), (100, 50), 45, 0, 360, (0, 255, 255), 5) # 중심 (600, 500), 반지름 (100, 50), 45도 회전, blue 타원
cv2.ellipse(img, (800, 500), (100, 50), -45, 0, 360, (0, 255, 255), 5) # 중심 (800, 500), 반지름 (100, 50), -45도 회전, blue 타원

plt.figure(figsize = (15, 15))
plt.imshow(img)
plt.show()
```

##### 사각형

> cv2.rectangle(이미지, 좌상단 좌표, 우하단 좌표, 색깔, 두께)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((480, 640, 3), dtype = np.uint8)

COLOR = (0, 255, 0) # BGR : Green
THICKNESS = 3

cv2.rectangle(img, (50, 150), (350, 400), (255, 255, 0)) # 속이 비워진 사각형, 두께 default 1
cv2.rectangle(img, (100, 100), (200, 200), COLOR, THICKNESS) # 속이 비워진 사각형
cv2.rectangle(img, (300, 100), (400, 300), COLOR, cv2.FILLED) # 속이 채워진 사각형

plt.figure(figsize = (8, 8))
plt.imshow(img)
plt.show()
```

##### 다각형

속이 비어있는 다각형

> cv2.polylines(이미지, 좌표(list로 감싸서 하나의 지점을 정의), 열림&닫힘(True&False), 색깔, 두께, 선 종류)

속이 채워진 다각형

> cv2.fillpoly(이미지, 좌표, 색깔, 선 종류)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((480, 640, 3), dtype = np.uint8)
img[:] = (255, 255, 255)

COLOR_1 = (255, 0, 0)
COLOR_2 = (0, 255, 0)
COLOR_3 = (0, 0, 255)
COLOR_4 = (255, 0, 255)

THICKNESS = 3

pts1 = np.array([[100, 100], [200, 100], [100, 200]]) # 세 점을 list로 정의
pts2 = np.array([[250, 150], [300, 75], [350, 100], [350, 200], [300, 225]])
pts3 = np.array([[400, 100], [500, 150], [400, 200]])
pts4 = np.array([[450, 150], [550, 100], [550, 200]])

# 속이 비어있는 다각형
cv2.polylines(img, [pts1], True, COLOR_1, THICKNESS, cv2.LINE_AA)
cv2.polylines(img, [pts2], True, COLOR_2, THICKNESS, cv2.LINE_AA)
cv2.polylines(img, [pts3, pts4], True, COLOR_3, THICKNESS, cv2.LINE_AA) # pts3, pts4 동시입력


pts5 = np.array([[[100, 300], [200, 300], [100, 400]], [[200, 300], [300, 300], [300, 400]]])

# 속이 채워진 다각형
cv2.fillPoly(img, pts5, COLOR_4, cv2.LINE_AA)

plt.figure(figsize = (8, 8))
plt.imshow(img)
plt.show()
```
